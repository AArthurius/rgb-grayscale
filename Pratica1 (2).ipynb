## Conversão de imagem RGB em imagem Grayscale

A ESCALA DE CINZA É O PROCESSO DE CONVERSÃO DE UMA IMAGEM DE OUTROS ESPAÇOS DE CORES, COMO RGB, CMYK, HSV, ETC., EM TONS DE CINZA. vARIANDO ENTRE PRETO E BRANCO. 

*A importância da escala de cinza para redução de dimensões (imagens em tons de cinza são unidimensionais), redução da complexidade do modelo e para que outros algoritmos funcionem (muitos algoritmos são personalizados para funcionarem em imagens em tons de cinza).*

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

Importando as bibliotecas pro código

*marplotlib -> biblioteca de plotagem para criar visualizações estáticas, animadas e interativas em python.*


```python
#abria a imagem
img = cv2.imread('data/t1.jpg',1)
cv2_imshow(img)
```

Abrindo a imagem colorida

*cv2.imread -> carrega a imagem do arquivo especificado. A imagem deve estar no diretório de trabalho ou um caminho completo da imagem deve ser fornecido.*

*vc2_imshow() é usado para exibir uma imagem em uma janela, esta, que se ajusta automaticamente ao tamanho da imagem.*

*waitKey(0) -> a função permite que os usuários exibam uma janela por determinados milissegundos ou até que qualquer tecla seja pressionada. Leva tempo em milissegundos como parâmetro e aguarda o tempo determinado para fechar a janela. Se for passado 0 no argumento ele espera até que qualquer tecla seja pressionada.*

```python
#aplicando conversão básica
#numpy
#img_grayscale_basic = (img[ : , : ,0]+img[ : , : ,1]+img[ : , : ,2])/3

#cv2
B, G, R = cv2.split(img)
img_grayscale_basic = (B+G+R)/3

img_grayscale_basic = np.array(img_grayscale_basic, dtype=np.uint8)
cv2_imshow(img_grayscale_basic)
#cv2.imshow('img_grayscale_basic', img_grayscale_basic)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

img_grayscale_basic = (B+G+R)/3 -> aplicando uma conversão de cores divindo o verde, azul e vermelho da imagem por três, que é uma forma muito básica de conversão para preto e branco.

*B, G, R = cv2.split(img) -> divisão da imagem multicanal em canais separados.*

*img_grayscale_basic = (B+G+R)/3 -> método que utiliza o valor médio de R (red), G(green) e B (blue) como o valor da escala de cinza. Pode haver algum erro de estouro se a soma de R, G e B for maior que 255. Para evitar, R, G e B devem ser calculados respectivamente (Escala de cinza = R/3 + G/3 + B/3).*

*img_grayscale_basic = np.array(img_grayscale_basic, dtype=np.uint8) -> Array*

*cv2_imshow(img_grayscale_basic)* -> exibe a imagem convertida.



```python
#aplicando conversão ponderada
#img_grayscale_basic = 0.299*img[ : , : ,0] + 0.587*img[ : , : ,1] + 0.114*img[ : , : ,2]

#cv2
B, G, R = cv2.split(img)
img_grayscale_pondered = 0.299*B+0.587*G+0.114*R

img_grayscale_pondered = np.array(img_grayscale_pondered, dtype=np.uint8)
cv2_imshow(img_grayscale_pondered)
#cv2.imshow('img_grayscale_basic', img_grayscale_pondered)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

img_grayscale_pondered = 0.299*B+0.587*G+0.114*R -> Os valores BGR são convertidos usando a formula padrão de conversão para preto e branco
